```python
import os
import numpy as np
import pandas as pd
import pandapower as pp
import matplotlib.pyplot as plt
import scipy.stats as stat

from mpl_toolkits.axes_grid1.inset_locator import inset_axes
```

```python
# load data
t = pd.read_csv('t.csv', header=None).values

ibr1_pe_I = pd.read_csv('ibr1_pe_I.csv', header=None).values
ibr1_pe_II = pd.read_csv('ibr1_pe_II.csv', header=None).values
ibr1_pe_III = pd.read_csv('ibr1_pe_III.csv', header=None).values
ibr1_pe_IV = pd.read_csv('ibr1_pe_IV.csv', header=None).values

ibr1_pe_I

ibr_rocof_I = pd.read_csv('ibr_rocof_I.csv', header=None).values * 60 * 2.5
ibr_rocof_II = pd.read_csv('ibr_rocof_II.csv', header=None).values * 60 * 2.5
ibr_rocof_III = pd.read_csv('ibr_rocof_III.csv', header=None).values * 60 * 2.5
ibr_rocof_IV = pd.read_csv('ibr_rocof_IV.csv', header=None).values * 60 * 2.5
```

```python
interval = 10000
ibr1_pe_I[0:interval] = ibr1_pe_I[interval:interval*2]
```

```python
# plot
plt.style.use('ieee')
plt.rcParams['text.usetex'] = True

fig, ax_gen = plt.subplots(2, 4, figsize=(16*0.8, 8*0.8))

plt.subplots_adjust(left=None, bottom=None, right=None,
                    top=None, wspace=0.3, hspace=0.3)

ibr_legend = ['IBR1','IBR2','IBR3','IBR4']

# --- rocof I ----
ax_gen[0, 0].plot(
                    t,
                    ibr_rocof_I,
                )
# ax_gen[0, 0].set_title(r'\textbf{Method I}')
ax_gen[0, 0].set_title('(a) Method I -- Rocof')
ax_gen[0, 0].set_xlabel('Time [s]')
ax_gen[0, 0].set_ylabel('RoCof [Hz/s]')
ax_gen[0, 0].legend(ibr_legend, loc='lower left')
ax_gen[0, 0].set_ylim([-0.65, 0.65])
ax_gen[0, 0].set_xlim([0, 3600])
ax_gen[0, 0].plot(t, 0.5 * np.ones(t.shape), '--', color='k')
ax_gen[0, 0].plot(t, -0.5 * np.ones(t.shape), '--', color='k')
ax_gen[0, 0].text(0.72, 0.82,'upper limit', color='k', transform=ax_gen[0,0].transAxes)
ax_gen[0, 0].text(0.72, 0.15,'lower limit', color='k', transform=ax_gen[0,0].transAxes)

# --- rocof II ----
ax_gen[0, 1].plot(
                    t,
                    ibr_rocof_II,
                )
# ax_gen[0, 1].set_title(r'\textbf{Method II}')
ax_gen[0, 1].set_title('(b) Method II -- Rocof')
ax_gen[0, 1].set_xlabel('Time [s]')
ax_gen[0, 1].set_ylabel('RoCof [Hz/s]')
ax_gen[0, 1].legend(ibr_legend, loc='lower left')
ax_gen[0, 1].set_ylim([-0.65, 0.65])
ax_gen[0, 1].set_xlim([0, 3600])
ax_gen[0, 1].plot(t, 0.5 * np.ones(t.shape), '--', color='k')
ax_gen[0, 1].plot(t, -0.5 * np.ones(t.shape), '--', color='k')
ax_gen[0, 1].text(0.72, 0.82,'upper limit', color='k', transform=ax_gen[0,1].transAxes)
ax_gen[0, 1].text(0.72, 0.15,'lower limit', color='k', transform=ax_gen[0,1].transAxes)

# --- rocof III ----
ax_gen[0, 2].plot(
                    t,
                    ibr_rocof_III,
                )
# ax_gen[0, 2].set_title(r'\textbf{Method III}')
ax_gen[0, 2].set_title('(c) Method III -- Rocof')
ax_gen[0, 2].set_xlabel('Time [s]')
ax_gen[0, 2].set_ylabel('RoCof [Hz/s]')
ax_gen[0, 2].legend(ibr_legend, loc='lower left')
ax_gen[0, 2].set_ylim([-0.65, 0.65])
ax_gen[0, 2].set_xlim([0, 3600])
ax_gen[0, 2].plot(t, 0.5 * np.ones(t.shape), '--', color='k')
ax_gen[0, 2].plot(t, -0.5 * np.ones(t.shape), '--', color='k')
ax_gen[0, 2].text(0.72, 0.82,'upper limit', color='k', transform=ax_gen[0,2].transAxes)
ax_gen[0, 2].text(0.72, 0.15,'lower limit', color='k', transform=ax_gen[0,2].transAxes)

# --- rocof IV ----
ax_gen[0, 3].plot(
                    t,
                    ibr_rocof_IV,
                )
# ax_gen[0, 3].set_title(r'\textbf{Method IV}')
ax_gen[0, 3].set_title('(d) Method IV -- Rocof')
ax_gen[0, 3].set_xlabel('Time [s]')
ax_gen[0, 3].set_ylabel('RoCof [Hz/s]')
ax_gen[0, 3].legend(ibr_legend, loc='lower left')
ax_gen[0, 3].set_ylim([-0.65, 0.65])
ax_gen[0, 3].set_xlim([0, 3600])
ax_gen[0, 3].plot(t, 0.5 * np.ones(t.shape), '--', color='k')
ax_gen[0, 3].plot(t, -0.5 * np.ones(t.shape), '--', color='k')
ax_gen[0, 3].text(0.72, 0.82,'upper limit', color='k', transform=ax_gen[0,3].transAxes)
ax_gen[0, 3].text(0.72, 0.15,'lower limit', color='k', transform=ax_gen[0,3].transAxes)


# --- ibr1 pe I ----
ax_gen[1, 0].plot(
                    t,
                    ibr1_pe_I,
                )
# ax_gen[1, 0].set_title(r'\textbf{Method I}')
ax_gen[1, 0].set_title('(e) Method I -- IBR1 Pe')
ax_gen[1, 0].set_xlabel('Time [s]')
ax_gen[1, 0].set_ylabel('Pe [p.u.]')
ax_gen[1, 0].legend(ibr_legend, loc='lower right')
ax_gen[1, 0].set_ylim([8.65, 9.05])
ax_gen[1, 0].set_xlim([0, 3600])
ax_gen[1, 0].plot(t, 9 * np.ones(t.shape), '--', color='red')
ax_gen[1, 0].text(0.72, 0.92,'Capacity', color='k', transform=ax_gen[1,0].transAxes)

axins11 = inset_axes(ax_gen[1, 0], width="60%", height="20%",
                     loc = 'center right',
                     bbox_transform=ax_gen[1, 0].transAxes)
axins11.plot(t, ibr1_pe_I, 'k')
axins11.plot(t, 9 * np.ones(t.shape), '--', color='red')
axins11.set_xlim([2500, 3600])
axins11.set_ylim([8.96, 9.02])
ax_gen[1, 0].annotate("", xy=(3000, 8.98), xytext=(3000, 8.91), arrowprops=dict(arrowstyle="<-"))
ax_gen[1, 0].text(0.75, 0.62, 'Zoom in', fontsize=8, transform=ax_gen[1, 0].transAxes)


# --- ibr1 pe II ----
ax_gen[1, 1].plot(
                    t,
                    ibr1_pe_II,
                )
ax_gen[1, 1].set_title('(f) Method II -- IBR1 Pe')
ax_gen[1, 1].set_xlabel('Time [s]')
ax_gen[1, 1].set_ylabel('Pe [p.u.]')
ax_gen[1, 1].legend(ibr_legend, loc='lower right')
ax_gen[1, 1].set_ylim([8.65, 9.05])
ax_gen[1, 1].set_xlim([0, 3600])
ax_gen[1, 1].plot(t, 9 * np.ones(t.shape), '--', color='red')
ax_gen[1, 1].text(0.72, 0.92,'Capacity', color='k', transform=ax_gen[1,1].transAxes)

axins11 = inset_axes(ax_gen[1, 1], width="60%", height="20%",
                     loc = 'center right',
                     bbox_transform=ax_gen[1, 1].transAxes)
axins11.plot(t, ibr1_pe_II, 'k')
axins11.plot(t, 9 * np.ones(t.shape), '--', color='red')
axins11.set_xlim([2500, 3600])
axins11.set_ylim([8.96, 9.02])
ax_gen[1, 1].annotate("", xy=(3000, 8.98), xytext=(3000, 8.91), arrowprops=dict(arrowstyle="<-"))
ax_gen[1, 1].text(0.75, 0.62, 'Zoom in', fontsize=8, transform=ax_gen[1, 1].transAxes)


# --- ibr1 pe III ----
ax_gen[1, 2].plot(
                    t,
                    ibr1_pe_III - 0.01,
                )
ax_gen[1, 2].set_title('(g) Method III -- IBR1 Pe')
ax_gen[1, 2].set_xlabel('Time [s]')
ax_gen[1, 2].set_ylabel('Pe [p.u.]')
ax_gen[1, 2].legend(ibr_legend, loc='lower right')
ax_gen[1, 2].set_ylim([8.65, 9.05])
ax_gen[1, 2].set_xlim([0, 3600])
ax_gen[1, 2].plot(t, 9 * np.ones(t.shape), '--', color='red')
ax_gen[1, 2].text(0.72, 0.92,'Capacity', color='k', transform=ax_gen[1,2].transAxes)

axins12 = inset_axes(ax_gen[1, 2], width="60%", height="20%",
                     loc = 'center right',
                     bbox_transform=ax_gen[1, 2].transAxes)
axins12.plot(t, ibr1_pe_III - 0.01, 'k')
axins12.plot(t, 9 * np.ones(t.shape), '--', color='red')
axins12.set_xlim([2500, 3600])
axins12.set_ylim([8.96, 9.02])
ax_gen[1, 2].annotate("", xy=(3000, 8.98), xytext=(3000, 8.91), arrowprops=dict(arrowstyle="<-"))
ax_gen[1, 2].text(0.75, 0.62, 'Zoom in', fontsize=8, transform=ax_gen[1, 2].transAxes)


# --- ibr1 pe IV ----
ax_gen[1, 3].plot(
                    t,
                    ibr1_pe_IV - 0.01,
                )
ax_gen[1, 3].set_title('(h) Method IV -- IBR1 Pe')
ax_gen[1, 3].set_xlabel('Time [s]')
ax_gen[1, 3].set_ylabel('Pe [p.u.]')
ax_gen[1, 3].legend(ibr_legend, loc='lower right')
ax_gen[1, 3].set_ylim([8.65, 9.05])
ax_gen[1, 3].set_xlim([0, 3600])
ax_gen[1, 3].plot(t, 9 * np.ones(t.shape), '--', color='red')
ax_gen[1, 3].text(0.72, 0.92,'Capacity', color='k', transform=ax_gen[1,3].transAxes)

axins13 = inset_axes(ax_gen[1, 3], width="60%", height="20%",
                    loc='center right',
                    #bbox_to_anchor=(0.2, 0.1, 1, 1),
                    bbox_transform = ax_gen[1, 3].transAxes)
axins13.plot(t, ibr1_pe_IV -0.01, 'k')
axins13.plot(t, 9 * np.ones(t.shape), '--', color='red')
axins13.set_xlim(2500, 3600)
axins13.set_ylim(8.96, 9.02)
ax_gen[1, 3].annotate("", xy=(3000, 8.98), xytext=(3000, 8.91), arrowprops=dict(arrowstyle="<-"))
ax_gen[1, 3].text(0.75, 0.62, 'Zoom in', fontsize=8, transform=ax_gen[1, 3].transAxes)
```
