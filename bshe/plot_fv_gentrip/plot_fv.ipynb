```python
import numpy as np
import pandas as pd
import pandapower as pp
import matplotlib.pyplot as plt
```

```python
plt.style.use('ieee')
plt.rcParams['text.usetex'] = True
```

```python
t_I   = np.loadtxt('t_I.csv', delimiter=',')
t_II  = np.loadtxt('t_II.csv', delimiter=',')
t_III = np.loadtxt('t_III.csv', delimiter=',')
t_IV  = np.loadtxt('t_IV.csv', delimiter=',')

f_I   = np.loadtxt('f_I.csv', delimiter=',')
f_II  = np.loadtxt('f_II.csv', delimiter=',')
f_III = np.loadtxt('f_III.csv', delimiter=',')
f_IV  = np.loadtxt('f_IV.csv', delimiter=',')

v_I   = np.loadtxt('busV_I.csv', delimiter=',')
v_II  = np.loadtxt('busV_II.csv', delimiter=',')
v_III = np.loadtxt('busV_III.csv', delimiter=',')
v_IV  = np.loadtxt('busV_IV.csv', delimiter=',')

rocof_I  = np.loadtxt('rocof_I.csv', delimiter=',')
rocof_II = np.loadtxt('rocof_II.csv', delimiter=',')
rocof_III= np.loadtxt('rocof_III.csv', delimiter=',')
rocof_IV = np.loadtxt('rocof_IV.csv', delimiter=',')
```

```python
fig_gen, ax_gen = plt.subplots(3, 4, figsize=(16*0.8/1.2, 8*0.8/2*3/1.2))

plt.subplots_adjust(left=None, bottom=None, right=None,
                    top=None, wspace=0.4, hspace=0.25)

legend = ['SG1', 'SG2', 'SG3', 'VSG1', 'VSG2', 'VSG3', 'VSG4', 'VSG5', 'VSG6', 'VSG7', 'VSG8']
legend_VSG = ['VSG1', 'VSG2', 'VSG3', 'VSG4', 'VSG5', 'VSG6', 'VSG7', 'VSG8']

# ---- I frequency ----
ax_gen[0, 0].plot(
                    t_I,
                    f_I * 60,
                    linewidth = 1,
                )
ax_gen[0, 0].set_title('Scheme I', fontweight='bold')
ax_gen[0, 0].set_xlabel('Time [s]')
ax_gen[0, 0].set_ylabel('Frequency [Hz]')
ax_gen[0, 0].legend(legend_VSG, loc='center right')
# ax_gen[0, 0].set_ylim([0.95, 1.12])
ax_gen[0, 0].set_xlim([0, 50])

# ---- II frequency ----
ax_gen[0, 1].plot(
                    t_II,
                    f_II * 60,
                    linewidth = 1,
                )
ax_gen[0, 1].set_title('Scheme II', fontweight='bold')
ax_gen[0, 1].set_xlabel('Time [s]')
ax_gen[0, 1].set_ylabel('Frequency [Hz]')
ax_gen[0, 1].set_xlim([0, 6.3])
ax_gen[0, 1].legend(legend_VSG, loc='center left')


# ---- III frequency ----
ax_gen[0, 2].plot(
                    t_III,
                    f_III * 60,
                    linewidth = 1,
                )
ax_gen[0, 2].set_title('Scheme III', fontweight='bold')
ax_gen[0, 2].set_xlabel('Time [s]')
ax_gen[0, 2].set_ylabel('Frequency [Hz]')
ax_gen[0, 2].set_xlim([0, 9])
ax_gen[0, 2].legend(legend_VSG, loc='center left')



# ---- IV frequency ----
ax_gen[0, 3].plot(
                    t_IV,
                    f_IV * 60,
                    linewidth = 1,
                )
ax_gen[0, 3].set_title('Scheme IV', fontweight='bold')
ax_gen[0, 3].set_xlabel('Time [s]')
ax_gen[0, 3].set_ylabel('Frequency [Hz]')
ax_gen[0, 3].set_xlim([0, 50])
ax_gen[0, 3].legend(legend_VSG, loc='center right')


## ------------------------- RoCoF --------------------------- ##
## ----------------------------------------------------------- ##

# ---- I rocof ----
ax_gen[1, 0].plot(
                    t_I,
                    rocof_I * 60,
                    linewidth = 1,
                )
ax_gen[1, 0].set_xlabel('Time [s]')
ax_gen[1, 0].set_ylabel('RoCoF [Hz/s]')
ax_gen[1, 0].set_xlim([0, 50])
ax_gen[1, 0].legend(legend_VSG)

# ---- II rocof ----
ax_gen[1, 1].plot(
                    t_II,
                    rocof_II * 60,
                    linewidth = 1,
                )
ax_gen[1, 1].set_xlabel('Time [s]')
ax_gen[1, 1].set_ylabel('RoCoF [Hz/s]')
ax_gen[1, 1].set_xlim([0, 6.3])
ax_gen[1, 1].legend(legend_VSG)

# ---- III rocof ----
ax_gen[1, 2].plot(
                    t_III,
                    rocof_III * 60,
                    linewidth = 1,
                )
ax_gen[1, 2].set_xlabel('Time [s]')
ax_gen[1, 2].set_ylabel('RoCoF [Hz/s]')
ax_gen[1, 2].set_xlim([0, 9])
ax_gen[1, 2].legend(legend_VSG)


# ---- IV rocof ----
ax_gen[1, 3].plot(
                    t_IV,
                    rocof_IV * 60,
                    linewidth = 1,
                )
ax_gen[1, 3].set_xlabel('Time [s]')
ax_gen[1, 3].set_ylabel('RoCoF [Hz/s]')
ax_gen[1, 3].set_xlim([0, 50])
ax_gen[1, 3].legend(legend_VSG)


## ------------------------- Voltage ------------------------- ##
## ----------------------------------------------------------- ##

# ---- I voltage ----
ax_gen[2, 0].plot(
                    t_I,
                    v_I,
                    linewidth = 0.5,
                )
ax_gen[2, 0].set_xlabel('Time [s]')
ax_gen[2, 0].set_ylabel('Bus Voltage [p.u.]')
ax_gen[2, 0].set_xlim([0, 50])

# ---- II voltage ----
ax_gen[2, 1].plot(
                    t_II,
                    v_II,
                    linewidth = 0.5,
                )
ax_gen[2, 1].set_xlabel('Time [s]')
ax_gen[2, 1].set_ylabel('Bus Voltage [p.u.]')
ax_gen[2, 1].set_xlim([0, 6.3])

# ---- III voltage ----
ax_gen[2, 2].plot(
                    t_III,
                    v_III,
                    linewidth = 0.5,
                )
ax_gen[2, 2].set_xlabel('Time [s]')
ax_gen[2, 2].set_ylabel('Bus Voltage [p.u.]')
ax_gen[2, 2].set_xlim([0, 9])

# ---- IV voltage ----
ax_gen[2, 3].plot(
                    t_IV,
                    v_IV,
                    linewidth = 0.5,
                )
ax_gen[2, 3].set_xlabel('Time [s]')
ax_gen[2, 3].set_ylabel('Bus Voltage [p.u.]')
ax_gen[2, 3].set_xlim([0, 50])

plt.savefig('fv_gentrip.png', dpi=600, bbox_inches='tight')
```
