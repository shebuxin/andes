```python
import torch
import torch.nn as nn
import numpy as np
```

```python
nn_name = 'trained_model.pth'

nn_para = torch.load(nn_name)
```

```python
nn_para
```

```python
nn_weight = []
nn_bias = []

items_list = list(nn_para.items())

for index, (key, value) in enumerate(items_list):
    if index % 2 == 0:
        nn_weight.append(value.numpy().astype(np.float64))
    else:
        nn_bias.append(value.numpy().astype(np.float64))
```
